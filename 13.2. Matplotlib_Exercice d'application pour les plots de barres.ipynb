
# Exercice d'étiquette de barre

```python
import matplotlib.pyplot as plt
import numpy as np
```

Définition des données

```python
N = 5
menMeans = (20, 35, 30, 35, -27)
womenMeans = (25, 32, 34, 20, -25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # les emplacements des x pour les groupes
width = 0.35       # la largeur des barres : peut aussi être la séquence len(x)
```

Diagramme à barres empilées avec barres d'erreur

```python
fig, ax = plt.subplots()

p1 = ax.bar(ind, menMeans, width, yerr=menStd, label='Men')
p2 = ax.bar(ind, womenMeans, width, bottom=menMeans, yerr=womenStd, label='Women')
#yerr cest l'erreur des données de y
#'bottom' poour s'etra stacked et au dessous(menMeans)
ax.axhline(0, color='grey', linewidth=0.8)
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind, labels=['G1', 'G2', 'G3', 'G4', 'G5'])
ax.legend()

# Label avec label_type 'center' au lieu du 'edge' par défaut
ax.bar_label(p1, label_type='center')
ax.bar_label(p2, label_type='center')
ax.bar_label(p2)
#thet chrtat f lwst fhmty
plt.show()
```

Diagramme à barres horizontales



```python
# Fixer l'état aléatoire pour la reproductibilité
np.random.seed(19680801)

people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

fig, ax = plt.subplots()

hbars = ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()  # étiquettes lues de haut en bas
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

# Étiquette avec des flotteurs spécialement formatés
ax.bar_label(hbars, fmt='%.2f')
ax.set_xlim(right=15)  # ajuster xlim pour s'adapter aux étiquettes

plt.show()
```

```python
fig, ax = plt.subplots()

hbars = ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()  
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

#Étiquette avec légendes manuelles, remplissage personnalisé et options d'annotation
ax.bar_label(hbars, labels=['±%.2f' % e for e in error],
             padding=20, color='b', fontsize=14)
ax.set_xlim(right=16)

plt.show()
#padding pour l'espaace entre les barres
```
