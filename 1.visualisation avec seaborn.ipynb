# visualisation avec seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
sns.set_theme()
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip")
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style='time') ;
#' style': pour le style d'affichage des scatter
#' hue 'variable qui produira des pionts avec une colonne with boollean et qui a plusieurs valeur (non boolean)
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",style='sex') 
## hu et style pour la memee colonne
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",size='size') 
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",size='size') 
## hue et style pour la memee colonne
```

```python
flights=sns.load_dataset('flights')
flights
```

```python
sns.lineplot(data=flights, x='year', y="passengers",estimator=max)
```

```python
sns.lineplot(data=flights, x='year', y="passengers",hue='month')
```

```python
rips=sns.load_dataset('taxis',parse_dates=["pickup","dropoff"])
rips
```

```python
rips['hour']=rips['pickup'].dt.hour
```

```python
rips
```

```python
sns.lineplot(data=rips, x='hour', y="total",hue='payment',style='color',ci=None)
#ci pour l'intervalle de confiance
```

```python
sns.relplot(data=rips, x='hour', y="total",kind='line',hue='payment',col='pickup_borough',row='dropoff_borough',ci=None);
# relplot() pour les sous figures
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=flights, x='year', y="passengers",hue='month')
```

```python
sns.relplot(data=tips, x='total_bill', y="tip",kind='scatter',hue='smoker',col='day',row='sex',height=3,aspect=1);
```

```python
sns.histplot(data=tips,x="tip",hue='smoker',multiple="dodge")
# dodge / stack / fill
```

```python
sns.histplot(data=tips,x="tip",hue='smoker')
```

```python
penguins=sns.load_dataset('penguins')
```

```python
penguins
```

```python
sns.histplot(data=penguins,x="body_mass_g",binwidth=500)
### pour la largeur de bins (binwidth)
```

```python
sns.histplot(data=penguins,x="body_mass_g",bins=20)
```

```python
sns.histplot(data=penguins,x="body_mass_g",bins=20,hue='species',multiple='stack',element='step')
#la forme d'aafichage du element du histogramme  'bar' est par dÃ©faut (poly// step)
```

```python
sns.histplot(data=penguins,x="body_mass_g",bins=20,hue='species',multiple='stack',palette='flare')
### plette les distributions du coleur
```

```python
sns.rugplot(data=tips,x='tip',height=0.1)
# rugplot() pour la creation descodebars
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
sns.rugplot(data=tips,x='total_bill',y='tip')
```

```python
sns.displot(kind='ecdf',data=penguins,x='body_mass_g',height=3,aspect=2)
# 'kde hist ecdf' les types de kind quise trouve dans displot()
```

```python
sns.displot(x="body_mass_g",data=penguins,col='species',hue='sex')
```

```python
sns.displot(data=tips,kind='hist',x='total_bill',y='tip',rug=True)
```

```python
sns.countplot(data=penguins,x='species',hue='sex')
```

```python
sns.countplot(data=tips,x='time',hue='smoker')
```

```python
sns.stripplot(data=rips,x='pickup_borough',y='distance')
plt.title('taxi trip distance by burough')
# stripplot() pour que les donnÃ©es ne seront plus chi dkhl f chi chauvauchement
```

```python
trip_sample=rips.nlargest(600,"total")
```

```python
plt.figure(figsize=(12,5))
sns.swarmplot(data=trip_sample,x='pickup_borough',y='distance')
```
