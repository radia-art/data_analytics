## Strip( )
Supprime les caractères de début et de fin.

Supprime les espaces (y compris les sauts de ligne) ou un ensemble de caractères spécifiés de chaque chaîne de la série/de l'index des côtés gauche et droit. Remplace toutes les non-chaînes de la série par des NaN.

```python
s = pd.Series(['1. Hawk.  ', '2. Pickle!\n', '3. Melonhead?\t'])
```

```python
s
```

```python
s.str.strip()
```

```python
s.str.strip(to_strip="123. \n \t")
```

```python
s.str.rstrip()
```

```python
s.str.lstrip()
```

```python
s.str.strip(to_strip="123.")
```

## Split( )
Sépare les chaînes autour d'un séparateur/délimiteur donné.

Fractionne la chaîne dans la série/l'index depuis le début, à la chaîne de délimitation spécifiée.

```python
titanic["home.dest"]
```

```python
titanic["home.dest"].str.split("/", expand=True)
```

```python
titanic["home"] = titanic["home.dest"].str.split("/", expand=True)[0]
```

```python
titanic
```

```python
titanic["destination"] = titanic["home.dest"].str.split("/", expand=True)[1]
```

```python
titanic
```

```python
titanic["home"].value_counts()
```

```python
titanic["destination"].value_counts()
```

**Remarque**:

**n** : Limite le nombre de fractionnements en sortie. None, 0 et -1 seront interprétés comme renvoyant tous les fractionnements.

```python
titanic["home.dest"].str.split("/", n=1, expand=True)
```

## Replace( ) 
Remplace chaque occurrence de motif/regex dans la série/l'index.

```python
ufos = pd.read_csv("nuforc_reports.csv")
```

```python
ufos
```

```python
ufos["duration"]
```

```python
ufos["duration"].str.replace("seconds", "s")
```

```python
ufos["duration"].str.replace("seconds|minutes", "", regex=True)
```

## Contains( )
Teste si un motif ou une expression régulière est contenu dans une chaîne d'une série ou d'un index.

Renvoie une série ou un index booléen en fonction du fait qu'un modèle ou une expression régulière donnés est contenu dans une chaîne d'une série ou d'un index.

```python
ufos["duration"].str.contains("hour")
```

```python
ufos["duration"]
```

```python
ufos[ufos["duration"].str.contains("day|week|month", na= False)]
#na : Remplit la valeur pour des valeurs manquantes.
```
