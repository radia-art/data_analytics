## Split( )
SÃ©pare les chaÃ®nes autour d'un sÃ©parateur/dÃ©limiteur donnÃ©.

Fractionne la chaÃ®ne dans la sÃ©rie/l'index depuis le dÃ©but, Ã  la chaÃ®ne de dÃ©limitation spÃ©cifiÃ©e.

```python
titanic["home.dest"]
```

```python
titanic["home.dest"].str.split("/", expand=True)
```

```python
titanic["home"] = titanic["home.dest"].str.split("/", expand=True)[0]
```

```python
titanic
```

```python
titanic["destination"] = titanic["home.dest"].str.split("/", expand=True)[1]
```

```python
titanic
```

```python
titanic["home"].value_counts()
```

```python
titanic["destination"].value_counts()
```

**Remarque**:

**n** : Limite le nombre de fractionnements en sortie. None, 0 et -1 seront interprÃ©tÃ©s comme renvoyant tous les fractionnements.

```python
titanic["home.dest"].str.split("/", n=1, expand=True)
```
