## Graphiques en boîte
Un box plot est une méthode permettant de représenter graphiquement des groupes de données numériques par le biais de leurs quartiles. La boîte s'étend des valeurs des quartiles Q1 à Q3 des données, avec une ligne à la médiane (Q2). Les moustaches partent des bords de la boîte pour montrer l'étendue des données. La position des moustaches est fixée par défaut à 1,5*IQR (IQR = Q3 - Q1) à partir des bords de la boîte. Les points extrêmes sont ceux qui dépassent l'extrémité des moustaches.

Lorsque vous utilisez ce graphique, il faut tenir compte du fait que la boîte et les moustaches peuvent se chevaucher, ce qui est très courant lorsque vous tracez de petits ensembles de données.

## Subplots
subplots bool ou séquence d'itérables, par défaut False

Si les colonnes doivent être regroupées en sous-plots :

- False : Aucun sous-plot ne sera utilisé

- True : Faire des sous-plots séparés pour chaque colonne.

sequence of iterables of column labels :

Crée un sous-graphe pour chaque groupe de colonnes. 

Par exemple [('a', 'c'), ('b', 'd')] créera 2 sous-plans : 

un avec les colonnes 'a' et 'c', et un avec les colonnes 'b' et 'd'. 

Les colonnes restantes qui ne sont pas spécifiées seront 

représentées dans des sous-plots supplémentaires (un par colonne). 

```python
df.plot(kind="hist")
```

```python
df.hist()
```

```python
df.plot(kind="hist", subplots=True, sharex=False, figsize=(5,5)) 
```

**Note**

**`sharexbool`**, Dans le cas où subplots=True, partage l'axe x et rend invisible les étiquettes de certains axes x ; la valeur par défaut est True si ax est None, sinon False si un axe est défini ; Attention, définir à la fois un axe et sharex=True modifiera toutes les étiquettes d'axe x pour tous les axes dans une figure.

**`plt.tight_layout()`** Permet d'ajuster les paramètres des sous-plots de façon à ce que les sous-plots soient bien adaptés à la figure. Ce faisant, seuls les étiquettes d'axe, les étiquettes de coche, les titres d'axe et les boîtes de décalage qui sont ancrées aux axes sont actuellement pris en compte.

```python
axs = df.plot(kind="hist", subplots=True,
              sharex=False, layout=(1,3), 
              figsize=(10,3), bins=30)
# Le tuple layout présente (lignes, colonnes) pour la mise en page des sous-plots.

plt.tight_layout()

axs[0][2].set_xlim(0,20000)
axs[0][1].set_title("Middle")
axs[0][0].set_title("First")
axs[0][2].set_title("Third")
```

## Sous-figures manuelles

```python
fig, axs = plt.subplots(1,2) # 1 seule ligne et 2 colonnes (axes des x et des y )
#plt.tight_layout()
ufos.year.value_counts().sort_index().plot(ax=axs[0])
axs[0].set_title("Sightings By Year")
ufos.month.value_counts().sort_index().plot(kind="bar", ax=axs[1])
axs[1].set_title("Sightings By Month")
```

```python
months = {1: 'Jan',
          2: 'Feb',
          3: 'Mar',
          4: 'Apr',
          5: 'May',
          6: 'Jun', 
          7: 'Jul',
          8: 'Aug',
          9: 'Sep',
          10: 'Oct',
          11: 'Nov',
          12: 'Dec'}
fig, axs = plt.subplots(2, 3, figsize=(14, 8)) #(2 lignes et 3 colonnes )
fig.suptitle("UFO sightings by month", fontsize=14)

fig.tight_layout(pad= 3) 
#pad : ajuste l'espace entre les sous-plots
ufos[ufos.year == 2014].month.value_counts().sort_index().rename(months).plot(kind="bar", ax=axs[0][0], title="2014")

ufos[ufos.year == 2015].month.value_counts().sort_index().rename(months).plot(kind="bar", ax=axs[0][1], title="2015")
ufos[ufos.year == 2016].month.value_counts().sort_index().rename(months).plot(kind="bar", ax=axs[0][2], title="2016")
ufos[ufos.year == 2017].month.value_counts().sort_index().rename(months).plot(kind="bar", ax=axs[1][0], title="2017")
ufos[ufos.year == 2018].month.value_counts().sort_index().rename(months).plot(kind="bar", ax=axs[1][1], title="2018")
ufos[ufos.year == 2019].month.value_counts().sort_index().rename(months).plot(kind="bar", ax=axs[1][2], title="2019")
```
