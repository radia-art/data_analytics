# Qu'est-ce que Pandas ?
Pandas est une bibliothèque Python utilisée pour travailler avec des ensembles de données.

Elle possède des fonctions d'analyse, de nettoyage, d'exploration et de manipulation des données.

Le nom "Pandas" fait référence à la fois à "Panel Data" et à "Python Data Analysis". Il a été créé par Wes McKinney en 2008.

# Pourquoi utiliser Pandas ?
Pandas nous permet d'analyser des données volumineuses et de tirer des conclusions fondées sur des théories statistiques.

Pandas peut nettoyer des ensembles de données désordonnés et les rendre lisibles et pertinents.

Les données pertinentes sont très importantes en science des données.

:}
La science des données : c'est une branche de l'informatique où l'on étudie comment stocker, utiliser et analyser les données pour en tirer des informations.

# Que peut faire Pandas ?
Voici quelques exemples de ce que pandas nous offre :

- Gestion facile des **données manquantes** (représentées par NaN, NA ou NaT) dans les données à virgule flottante et non flottante.
- Mutabilité de la taille : des colonnes peuvent être **insérées et supprimées ** de DataFrame et d'objets de dimension supérieure
- **Alignement automatique et explicite des données** : les objets peuvent être alignés explicitement sur un ensemble d'étiquettes, ou l'utilisateur peut simplement ignorer les étiquettes et laisser Series, DataFrame, etc. aligner automatiquement les données pour vous dans les calculs.
- Fonctionnalité de **regroupement** puissante et flexible permettant d'effectuer des opérations de fractionnement, d'application et de combinaison sur des ensembles de données, tant pour l'agrégation que pour la transformation des données.
- **Facilite la conversion** en objets DataFrame de données fragmentées et indexées différemment dans d'autres structures de données Python et NumPy.
- Découpage intelligent basé sur **les étiquettes, indexation élégante et sous-ensembles** de grands ensembles de données.
- **Fusion et jonction** intuitives d'ensembles de données
- Remodelage et pivotement flexibles des ensembles de données
- **Étiquetage hiérarchique** des axes (possibilité d'avoir plusieurs étiquettes par tick)
- Outils IO robustes pour le chargement de données à partir de **fichiers plats** (CSV et délimités), de **fichiers Excel, de bases de donnée**s et pour l'enregistrement/le chargement de données à partir du format **HDF5 ultrarapide.
- Fonctionnalités spécifiques aux **séries temporelles**: génération de plages de dates et conversion de fréquences, statistiques de fenêtres mobiles, décalage et retard de dates.

# Quel type de données les pandas traitent-ils ?
Lorsque on travaille avec des données tabulaires, telles que des données stockées dans des feuilles de calcul ou des bases de données, pandas est l'outil qu'il vous faut. pandas vous aidera à explorer, nettoyer et traiter vos données. 

# Structures de données de Pandas
Pour commencer, vous devez charger le paquet Pandas via son alias ***pd*** comme convenu par la communauté Pandas. Vous devez également importer NumPy, puisque NumPy et Pandas sont deux bibliothèques qui se complètent.

```python
import pandas as pd
import numpy as np
```

On distingue deux structures de données Pandas :
- **Series** : est un tableau étiqueté unidimensionnel pouvant contenir tout type de données (entiers, chaînes de caractères, nombres à virgule flottante, objets Python, etc.) Les étiquettes des axes sont appelées index. 
- **DataFrame** :  est une structure de données étiquetée en deux dimensions avec des colonnes de types potentiellement différents. C'est généralement l'objet pandas le plus utilisé.


**Remarque** : Notez que chaque colonne d'un DataFrame est une série.
