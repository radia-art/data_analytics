# Conversion en type date : ``pd.to_datetime( )``

Cette fonction convertit l'argument que ce soit un scalaire, un tableau, une série ou un DataFrame/dict en un objet pandas datetime(date/heure).

```python
import pandas as pd
```

```python
pd.to_datetime("2022-12-12")
```

```python
pd.to_datetime("2022/12/12")
```

```python
pd.to_datetime("December 12th 2022")
```

```python
pd.to_datetime("December 12th 2022 4pm")
```

```python
pd.to_datetime("December 12 2022 4:50am")
```

```python
pd.to_datetime("Dec. 12 2022 12pm")
```

```python
pd.to_datetime("3:55:45 Mar. 25th 94")
```

```python
dates = ["3:55:45 Apr. 3rd 90", 
         "3:55:45 Apr. 22nd 91", 
         "7:55:45 Jan. 21st 90"]
pd.to_datetime(dates)
```

```python
pd.to_datetime("13/12/22")
```

```python
pd.to_datetime("13/12/22", dayfirst=True) 
# "dayfirst" bool, par défaut False
#Spécifiez un ordre d'analyse de date si arg est str ou ressemble à une liste. 
#Si True, analyse les dates avec le jour en premier, 
#par ex. "10/11/12" est analysé comme 2012-11-10.
```

```python
pd.to_datetime("13/12/22", yearfirst=True)
#"yearfirst" bool, par défaut False
#Spécifiez un ordre d'analyse de date si arg est str ou ressemble à une liste.

#If True analyse d'abord les dates avec l'année,
# par exemple. "10/11/12" est analysé comme 2010-11-12.
```

```python
pd.to_datetime("12/10/22", yearfirst=True, 
               dayfirst=True) 
# Si dayfirst et yearfirst sont tous deux True, yearfirst est précédé
```

```python
pd.to_datetime("10/11/12", format="%d/%y/%m")
# "formatstr", par défaut Aucun
# Le strftime pour analyser le temps,
# par exemple. "%d/%m/%A".
```

```python
meetings = ["Dec 11 2019 Meeting",
            "Jan 6 2020 Meeting", 
            "Mar 19 2020 Meeting", 
            "Apr 15 2020 Meeting"]
```

```python
pd.to_datetime(meetings, 
               format="%b %d %Y Meeting")
```
